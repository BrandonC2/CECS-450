```python
from copy import deepcopy
d = {'name' : ['Alfred', 'Alex']}
c = d.copy() # any change to d dictionary, will be reflected; objects only
cd = deepcopy(d) # now d and f dictionary are separate

d['name'].append('John')
c['name'].append('Steward')
cd['name'].append('Mike')


print(d)
print(c)
print(cd)
```

```python
d['name'].remove('John')
print(d)
```

```python
def my_calculator(x, y, operation = None):
    if operation is None:
        return
    elif operation == 'sum':
        return x + y
    elif operation == 'sub':
        return x - y
    elif operation == 'mult':
        return x * y
    elif operation == 'div':
        return x / y
    else:
        raise ValueError('Unknown operation')
    
print(my_calculator(2,3, operation = 'mult'))
```

```python
# you have a DNA sequence
# define a function that counts the bases
# input should be a string
# check if it's an DNA

DNA = "AAGCCCTGAGAGCC"

def DNA_counter(DNA):
    if DNA is None:
        raise ValueError("Please enter a valid sequence")
    base_count = {'A': 0, 'G': 0, 'C': 0, 'T': 0}
    count = 0

    while count < len(DNA):
        if DNA[count] == 'A':
            base_count['A'] += 1
        elif DNA[count] == 'G':
            base_count['G'] += 1
        elif DNA[count] == 'C':
            base_count['C'] += 1
        elif DNA[count] == 'T':
            base_count['T'] += 1
        else:
            raise ValueError("Invalid DNA sequence")
        count += 1  


    return base_count


print(DNA_counter(DNA))

```

```python
seq = "AAGCCCTGAGAGCC"

def Count_DNA(seq):
    if seq is None:
        raise ValueError("Please enter a valid sequence")

    Counts = {}
    countA = 0
    countG = 0
    countC = 0
    countT = 0

    for i in range(len(seq)):
        if seq[i] == 'A':
            countA += 1
        elif seq[i] == 'G':
            countG += 1
        elif seq[i] == 'C':
            countC += 1
        elif seq[i] == 'T':
            countT += 1
        else:
            raise ValueError("Error: This is not a DNA sequence")

    Counts['A'] = countA
    Counts['G'] = countG
    Counts['C'] = countC
    Counts['T'] = countT

    return Counts

print(Count_DNA(seq))
```
