```python
class MyClass:
    variable = 'Hello' # only defined inside the class; local varaible

    def my_function(self): # join all the function and class to each other?
        print("This is a message inside the class")

my_object = MyClass()

print(my_object.variable) # . refers to something inside the class

print(MyClass().variable)

my_object.my_function() # runs the function inside the class
```

```python
class Person:
    def __init__(self, name = None, age = None): # need init for variables to be global
        self.name = name # self allows name to be global
        self.age = age
        self.school = "CSULB" # fixed variable

    def ethnicity(self, e):
        print(self.name) # name is now access by other functions through "self."
        self.e = e
        if self.name is None:
            self.name = input("Please enter a name for the person: ")
        elif self.age is None:
            self.age = input("Please entere an age for the person: ")

        print("My name is ", self.name, ". My age is ", self.age, ". My ethnicity is ", self.e)


p1 = Person(age = 36)
p2 = Person("Jack", 54)
p3 = Person()

p1.ethnicity("White")
p2.ethnicity("Black")
```

```python
class Bag:
    def __init__(self):
        self.data = []
    
    def add(self, x):
        self.data.append(x)

    def addtwice(self, x):
        self.add(x)
        self.add(x)

    def add_n_times(self, x, n):
        for i in range(n):
            self.add(x)

My_Bag = Bag()
My_Bag.add(1)
My_Bag.add_n_times(2, n=3)
My_Bag.add([4,5])
print(My_Bag.data)
```

```python
class my_sum:
    def add_f(self, a, b):
        return a + b

class math_op:
    def sum_three_in(self, a, b, c):
        x = my_sum().add_f(a,b)
        y = my_sum().add_f(x,c)
        return y
    
print(math_op().sum_three_in(1,2,3))
    

```
